## Semi-Supervised Techniques

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.semi_supervised import LabelPropagation
from sklearn.semi_supervised import LabelSpreading
from methods import scikitTSVM
from scipy import linalg
from sklearn.decomposition import PCA


# Generative model implementing knn clustering for the prediction of the labels of the unlabeled samples
def knn_ssl(k, lab_data, y, unlab_data, yy, num_unlab):
    
    # Extracting the means of the labeled samples for the initialization of the clusters of KMeans
    ind_g = y=='g'
    ind_h = y=='h'
    mean_g = np.mean(lab_data[ind_g,:], 0)
    mean_g = mean_g[np.newaxis, :]
    mean_h = np.mean(lab_data[ind_h,:], 0)
    mean_h = mean_h[np.newaxis, :]
    
    label0 = 'g'
    label1 = 'h'
    
    # Initializing the clustering method with the extracted means
    kmeans = KMeans(n_clusters=k, init = np.append(mean_g, mean_h, 0),  n_init=1, n_jobs=-1)
    
    # Random sampling of unlabeled data and re-fitting the Kmeans clusters
    ii = np.floor(np.random.rand(num_unlab) * len(unlab_data)).astype(int)
    udata = unlab_data[ii,:]
    kmeans.fit(udata)
    pred = kmeans.labels_
    clabels = pred.astype(str)
    clabels[clabels=='0'] = label0
    clabels[clabels=='1'] = label1
    
    # Estimating the clustering error with respect to the actual data
    cle = np.sum(clabels != yy[ii])/len(ii)
    
    # Labeled and unlabeled data are stored in a single array and shuffled
    data = np.append(lab_data, udata, axis=0)
    labels = np.append(y, clabels)
    ind = np.arange(len(labels))
    np.random.shuffle(ind)
    
    return data[ind,:], labels[ind], cle

# Self-learning  method, where LDA is also used for the prediction of the labels of the unlabelled samples. Only the most confident
# samples are added. No retraining
def lda_ssl(lab_data, y, unlab_data, yy, num_unlab, threshold):
    
    # Initial fitting of the LDA using the labelled samples
    clf = LDA()
    clf.fit(lab_data, y)
    #initializing these two matrices which will be returned as the new sets of labelled samples. Only the most confident predictions
    # are kept and the classifier is retrained in order to predict the left out unlabelled samples.
    data = lab_data
    labels = y
    ss = 0
    
    #random sampling of the unlabeled samples from the original set
    ii = np.floor(np.random.rand(num_unlab) * len(unlab_data)).astype(int)
    udata = unlab_data[ii,:]
    ulabels = yy[ii]
    
    # The self learning classifier is retrained with the most confident samples and repredicts. If confidents samples are found, they
    # are added to the set. If not the function returns the set.
    
    pred = clf.predict(udata)
    conf = np.amax(clf.predict_proba(udata), axis=1)
    ui = conf>threshold    
    
    if np.sum(ui)==0:
        ind = np.arange(len(labels))
        np.random.shuffle(ind)
        return data[ind,:], labels[ind], ss
    else:
        # extracting the error of the semi-supervised LDA for the prediction of the labels for the unlabeled samples
        ss = np.sum(pred[ui] != ulabels[ui])/len(ui)
        # update the set of the labelled samples
        data = np.append(lab_data, udata[ui,:], axis=0)
        labels = np.append(y, pred[ui])
        # Shuffling to infer variance
        ind = np.arange(len(labels))
        np.random.shuffle(ind)
        return data[ind,:], labels[ind], ss
    

# Self-learning  method, where LDA is also used for the prediction of the labels of the unlabelled samples. Retraining occurs as an iterative
# process with the most confident samples.
def lda_ssl_iter(lab_data, y, unlab_data, yy, num_unlab, threshold):
    
    # Initial fitting of the LDA using the labelled samples
    clf = LDA()
    clf.fit(lab_data, y)
    #initializing these two matrices which will be returned as the new sets of labelled samples. Only the most confident predictions
    # are kept and the classifier is retrained in order to predict the left out unlabelled samples.
    data = lab_data
    labels = y
    
    #random sampling of the unlabeled samples from the original set
    ii = np.floor(np.random.rand(num_unlab) * len(unlab_data)).astype(int)
    udata = unlab_data[ii,:]
    ulabels = yy[ii]
    
    # The self learning classifier is retrained with the most confident samples and repredicts. If confidents samples are found, they
    # are added to the set. If not the function returns the set.
    c = 0
    ss = 0
    while True:
        pred = clf.predict(udata)
        conf = np.amax(clf.predict_proba(udata), axis=1)
        ui = conf>threshold
        #print(np.sum(ui))
        
        if np.sum(ui)==0:
            ind = np.arange(len(labels))
            np.random.shuffle(ind)
            break
        else:
            # extracting the error of the semi-supervised LDA for the prediction of the labels for the unlabeled samples
            ss = np.sum(pred[ui] != ulabels[ui])/len(ui)
            # update the set of the labelled samples
            data = np.append(lab_data, udata[ui,:], axis=0)
            labels = np.append(y, pred[ui])
            # Shuffling to infer variance
            ind = np.arange(len(labels))
            np.random.shuffle(ind)
            # updating the set of the unlabelled samples which are going to be repredicted
            udata = udata[conf<=threshold]
            if len(udata)==0:
                break
            ulabels = ulabels[conf<=threshold]
            # retrain and repredict
            clf.fit(data[ind,:], labels[ind])
            c+=1
            
    #print(c)
    return data[ind,:], labels[ind], ss
    

# This method uses label propagation for the prediction of the labels of the unlabeled samples.
def lprop_ssl(lab_data, y, unlab_data, yy, num_unlab):
    y_int = np.zeros(len(y))
    y_int[y=='h'] = 1
    
    clf = LabelPropagation(kernel = 'rbf', gamma=20, max_iter=2000, n_jobs = -1)
    ii = np.floor(np.random.rand(num_unlab) * len(unlab_data)).astype(int)
    udata = unlab_data[ii,:]
    
    pred = -1 * np.ones(num_unlab)
    data = np.append(lab_data, udata, axis=0)
    labels = np.append(y_int, pred)
    
    ind = np.arange(len(labels))
    np.random.shuffle(ind)

    clf.fit(data[ind,:], labels[ind])
    lab = np.zeros(len(labels), dtype=str)
    lab[clf.transduction_==0] = 'g'
    lab[clf.transduction_==1] = 'h'
    
    return data[ind,:], lab


def constrained_lda(lab_data, y, unlab_data, yy, num_unlab):
    
    
    ind_g = y=='g'
    ind_h = y=='h'
    
    ii = np.floor(np.random.rand(num_unlab) * len(unlab_data)).astype(int)
    udata = unlab_data[ii,:]
    ulabels = yy[ii]
    
    clf = LDA(store_covariance=True)
    clf.fit(lab_data, y)
    
    # Extracting the class priors of the labelled sample set
    priors0 = np.sum(ind_g)/len(y)
    priors1 = np.sum(ind_h)/len(y)
    tau = clf.covariance_ 
    mean = clf.priors_[0] * clf.means_[0,:] + clf.priors_[1] * clf.means_[1,:] 

    pred = clf.predict(udata)
    data = np.append(lab_data, udata, 0)
    labels = np.append(y, ulabels, 0)
    
    
    clf.fit(data, labels)
    theta = clf.covariance_ 
    mhu = clf.priors_[0] * clf.means_[0,:] + clf.priors_[1] * clf.means_[1,:]
    
    tfdata = np.matmul(np.dot(linalg.sqrtm(theta), linalg.sqrtm(np.linalg.inv(tau))), data.T - mean[:,np.newaxis]) + mhu[:, np.newaxis]
    tfdata = tfdata.T
    
    ind = np.arange(len(labels))
    np.random.shuffle(ind)
    return tfdata[ind,:], labels[ind], 0

def whiten(X):

    U, s, Vt = np.linalg.svd(X, full_matrices=False)
    wX = np.dot(U, Vt)
    return wC
```

## Plotting curves functions

```python
np.seterr(over = 'ignore', under = 'ignore')
print(np.geterr())
def plot_learning_curves(train_data, y, test_data, yy, rep, num_unlab, num_lab):
    
    mean_true_error = np.zeros(len(num_unlab))
    mean_apparent_error = np.zeros(len(num_unlab))
    std_true_error = np.zeros(len(num_unlab))
    std_apparent_error = np.zeros(len(num_unlab))
    mean_ss_error = np.zeros(len(num_unlab))
    std_ss_error = np.zeros(len(num_unlab))
    
    a_error = np.zeros(rep)
    t_error = np.zeros(rep)
   
    clf = LDA()
    
    for j in range(len(num_unlab)):
        sss = StratifiedShuffleSplit(n_splits=rep, test_size= None, train_size = num_lab)
        i=0
        ss = np.zeros(rep)
        for lab_ind, unlab_ind in sss.split(train_data, y):
            if  num_unlab[j]==0:
                clf.fit(train_data[lab_ind,:], y[lab_ind])                               
                preda = clf.predict(train_data[lab_ind,:])
                predt = clf.predict(test_data)
                a_error[i] = np.sum(preda != y[lab_ind])/len(preda)
                t_error[i] = np.sum(predt != yy)/len(predt)
            else:
                #dssl, lssl, ss[i] = constrained_lda(train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j])
                dssl, lssl, ss[i] = knn_ssl(2, train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j])
                #dssl, lssl, ss[i] = lda_ssl(train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j], 0.95)
                #dssl, lssl, ss[i] = lda_ssl_iter(train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j], 0.95)
                #dssl, lssl = lprop_ssl(train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j])
                #dssl, lssl = lspread_ssl(train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j])
                clf.fit(dssl, lssl)
                preda = clf.predict(dssl)
                predt = clf.predict(test_data)
                a_error[i] = np.sum(preda != lssl)/len(preda)
                t_error[i] = np.sum(predt != yy)/len(predt)
                #print(a_error[i], t_error[i])
            i += 1
        mean_true_error[j] = np.mean(t_error)
        mean_apparent_error[j] = np.mean(a_error)
        std_true_error[j] = np.std(t_error)
        std_apparent_error[j] = np.std(a_error)
        mean_ss_error[j] = np.mean(ss)
        std_ss_error[j] = np.std(ss)
        
    print(mean_true_error)
    fig = plt.figure(figsize=(8,4))
    plt.title('Learning Curves vs #unlabeled samples')
    plt.xlabel("# of unlabeled samples")
    plt.ylabel("Error rate")
    aa = np.arange(len(num_unlab))
    plt.fill_between(aa, mean_true_error - std_true_error,
                     mean_true_error + std_true_error, alpha=0.1, color="r")
    plt.plot(aa, mean_true_error, 'o-', color="r", label = 'True Error')
    #plt.fill_between(aa, mean_apparent_error - std_apparent_error,
     #               mean_apparent_error + std_apparent_error, alpha=0.1, color="g")
    #plt.plot(aa, mean_apparent_error, 'o-', color="g", label = 'Apparent Error')
    plt.fill_between(aa, mean_ss_error - std_ss_error,
                    mean_ss_error + std_ss_error, alpha=0.1, color="b")
    plt.plot(aa, mean_ss_error, 'o-', color="b", label = 'SS Error')
    plt.axis([0, 1, 0, 1], aspect = 'equal')
    
    labels = ['0', '10', '20', '40', '80', '160', '320', '640', '1280', '2560']
    plt.xticks(aa, labels, rotation = 'horizontal')
    plt.grid()
    plt.legend(loc="best")
    
    return plt

def plot_ll_curves(train_data, y, test_data, yy, rep, num_unlab, num_lab):
    
    clf = LDA()
    mean_LL = np.zeros(len(num_unlab))
    std_LL = np.zeros(len(num_unlab))
    mean_ss_error = np.zeros(len(num_unlab))
    std_ss_error = np.zeros(len(num_unlab))
    
    for j in range(len(num_unlab)):
        LL = np.zeros(rep)
        ss = np.zeros(rep)
        sss = StratifiedShuffleSplit(n_splits=rep, test_size = None, train_size = num_lab)
        i=0
        for lab_ind, unlab_ind in sss.split(train_data, y):
            if  num_unlab[j]==0:
                clf.fit(train_data[lab_ind,:], y[lab_ind])
                log_prob = clf.predict_proba(test_data)
                log_prob = np.nan_to_num(log_prob)
                log_prob = np.amax(log_prob, axis=1)
                #print(log_prob.shape)
                LL[i] = np.sum(np.log(log_prob))
            else:
                #dssl, lssl, ss[i] = constrained_lda(train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j])
                dssl, lssl, ss[i] = knn_ssl(2, train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j])
                #dssl, lssl, ss[i] = lda_ssl(train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j], 0.95)
                #dssl, lssl, ss[i] = lda_ssl_iter(train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j], 0.95)
                #dssl, lssl = lprop_ssl(train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j])
                #dssl, lssl = lspread_ssl(train_data[lab_ind,:], y[lab_ind], train_data[unlab_ind,:], y[unlab_ind], num_unlab[j])
                clf.fit(dssl, lssl)
                log_prob = clf.predict_proba(test_data)
                log_prob = np.nan_to_num(log_prob)
                log_prob = np.amax(log_prob, axis=1)
                LL[i] = np.sum(np.log(log_prob))
            i += 1
        mean_LL[j] = np.mean(LL)
        std_LL[j] = np.std(LL)
        mean_ss_error[j] = np.mean(ss)
        std_ss_error[j] = np.std(ss)
        

    fig = plt.figure(figsize=(8,4))
    plt.title('Log-Likelihood Curve vs #unlabeled samples')
    plt.xlabel("#unlabeled samples")
    plt.ylabel("LL")
    aa = np.arange(len(num_unlab))
    plt.fill_between(aa, mean_LL - std_LL,
                    mean_LL + std_LL, alpha=0.1, color="b")
    plt.plot(aa, mean_LL, 'o-', color="b", label = 'Log-Likelihood')
    #plt.axis([0, 1, -5, 1], aspect = 'equal')
    
    labels = ['0', '10', '20', '40', '80', '160', '320', '640', '1280', '2560']
    plt.xticks(aa, labels, rotation = 'horizontal')
    plt.grid()
    plt.legend(loc="best")
    
    return plt
```
