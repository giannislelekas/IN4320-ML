# Importing Data

```python
import numpy as np
from matplotlib import pyplot as plt
import csv
import pandas as pd

train_file = "/home/giannislelekas/git/Machine_Learning/IN4320-ML/data/train.csv"
test_file = "/home/giannislelekas/git/Machine_Learning/IN4320-ML/data/test.csv"
label_file = "/home/giannislelekas/git/Machine_Learning/IN4320-ML/data/label.csv"

# Read the train.csv and store train_data and train_labels np array
df = pd.read_csv(train_file, header=None)
train_data = df.values
train_labels = train_data[:, 0]
train_data = train_data[:, 1:]

# Read the test.csv and store test_data and test_labels np array
df = pd.read_csv(test_file, header=None)
test_data = df.values

print("Size of the training data:", train_data.shape)
print("Size of the test data:", test_data.shape)
```

## Exploring the data

```python
plt.figure()
plt.hist(train_labels, bins=3)
plt.title("Class priors")

print("NaN values in training:", np.sum(np.isnan(train_data)))
print("NaN values in testing:", np.sum(np.isnan(test_data)))

# Standard deviation
f, ax = plt.subplots(1,2)
f.set_figwidth(20)
ax[0].set_title("std in train_data")
ax[0].set_xlabel("dim")
ax[0].plot(range(train_data.shape[1]), np.std(train_data, axis=0))
ax[1].set_title("std in test_data")
ax[1].set_xlabel("dim")
ax[1].plot(range(test_data.shape[1]), np.std(test_data, axis=0))

# Max values
f, ax = plt.subplots(1,2)
f.set_figwidth(20)
ax[0].set_title("max in train_data")
ax[0].set_xlabel("dim")
ax[0].plot(range(train_data.shape[1]- 1), np.amax(train_data[:, 1:], axis=0))
ax[1].set_title("max in test_data")
ax[1].set_xlabel("dim")
ax[1].plot(range(test_data.shape[1]-1), np.amax(test_data[:, 1:], axis=0))

# Min values
f, ax = plt.subplots(1,2)
f.set_figwidth(20)
ax[0].set_title("min in train_data")
ax[0].set_xlabel("dim")
ax[0].plot(range(train_data.shape[1]- 1), np.amin(train_data[:, 1:], axis=0))
ax[1].set_title("max in test_data")
ax[1].set_xlabel("dim")
ax[1].plot(range(test_data.shape[1]-1), np.amin(test_data[:, 1:], axis=0))
```

## Classification

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn import svm

n_splits = 10
cval = StratifiedShuffleSplit(n_splits= n_splits, test_size=0.5)
clf = svm.SVC(C=1.0, kernel='rbf', degree=3, gamma='auto', coef0=0.0, shrinking=True, probability=False, 
              tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1, 
              decision_function_shape='ovr', random_state=None)

app_error = np.zeros(n_splits)
app_error2 = np.zeros(n_splits)
true_error = np.zeros(n_splits)
true_error2 = np.zeros(n_splits)

c=0
for train_index, test_index in cval.split(train_data, train_labels):
    clf.fit(train_data[train_index], train_labels[train_index])
    app_error[c] = 1 - clf.score(train_data[train_index], train_labels[train_index])
    pred = clf.predict(train_data[train_index])
    app_error2[c] = np.sum(pred != train_labels[train_index])/len(train_labels[train_index])
    true_error[c] = 1 - clf.score(train_data[test_index], train_labels[test_index])
    pred = clf.predict(train_data[test_index])
    true_error2[c] = np.sum(pred != train_labels[test_index])/len(train_labels[test_index])
    c+=1
```

```python
plt.figure
plt.plot(range(n_splits), app_error,'g')
plt.plot(range(n_splits), true_error, 'r')
```
